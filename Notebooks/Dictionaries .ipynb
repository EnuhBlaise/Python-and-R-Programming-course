# Definition
Dictionaries are used to store information that have a relation for a example a name and phone number.
These are stored in Key value relationships 
 (image from pythontic.com)

```python
Dict = {}
Dict = {"Kim": "Istanbul University", "Ali": "Osmangazi University", "Ismael":"Konya University"}
print(Dict) #printing the entire dictionary
print(Dict["Kim"])#printing a value in the dictionary
```

```python
#*Type of Values in Dictionaries*
#Python allows the values in a dictionary to be any type – string, integer, a list, another dictionary, boolean, etc. 
#However, keys must always be an immutable data type, such as strings, numbers, or tuples. 

Dict = {"Kim": "Istanbul University", #string and String
        "Age": 26, #String and integer combined
        "Favourite movies": ["Avengers", "Fisherman's Diary", "Annabel"],#String and Lists combined
        "Friend's Information": {"Sam": 26, "Steph": 35, "Ali": 48} #Nested Dictionaries
       }
```

```python
Dict["Friend's Information"]
```

# Iterating Dictionary Items

```python
#iteration with a for loop (Iteration is possible over keys and Values)
for keys in Dict.keys(): #Dict.values() for values
    print(keys)
```

```python
for keys in Dict_1.keys(): #Dict.values() for values
    print(keys)
```

```python
#iteration over both keys and values 
for items in Dict.items(): #Dict.values() for values
    print(items)  #returns response as tuples.
```

# Dictionary comprehension

```python
sq_numbers = {}
for a in range(5):
    sq_numbers[a] = a**2
    
print(sq_numbers)
```

```python
#With Dictionary comphrehensions
sq_numbers_2 = {x:x**2 for x in range(5)}
print(sq_numbers_2)
```

```python
#Two lists can be zip into a dictionary
List1 = ["a", "b", "c"]
List2 = ["x", "y", "z"]

List3 = zip(List1, List2)#forms an iterable object which can be acessed with a for loop
```

```python
{x:y for x,y in List3}
```
